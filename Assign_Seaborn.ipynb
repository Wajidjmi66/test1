# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

## Answer :->
## Seaborn is a versatile library for statistical data visualization in Python. Here are five commonly used plots along with their uses:

# 1. Scatter Plot:
### Use: Shows the relationship between two numerical variables.
### Purpose: Visualize patterns, correlations, or clusters in data points.

# 2. Bar Plot:
### Use: Displays categorical data with rectangular bars.
### Purpose: Compare categories or show counts/frequencies for different groups.

# 3. Histogram:
### Use: Represents the distribution of a single numerical variable.
### Purpose: Illustrate the frequency or probability distribution of data.

# 4. Line Plot:
### Use: Shows data points connected by straight lines.
### Purpose: Display trends or changes over a continuous variable (often time).

# 5. Box Plot (or Box-and-Whisker Plot):
### Use: Summarizes the distribution of a numerical variable.
### Purpose: Identify outliers, visualize median, quartiles, and range in the data.
#### These plots serve various purposes and are useful for exploring, understanding, and communicating insights from datasets.








# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
## Note: timepoint, signal, event, and region are columns in the fmri dataset.

## Answer:->

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")
```

```python
fmri_data
```

```python
# Create a line plot using Seaborn
plt.figure(figsize=(8, 4))  # Adjust figure size if needed

# Plot line plot for different events and regions
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and labels
plt.title('Signal Change over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)  # Adjust legend position
plt.tight_layout()
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
## Note: pclass, age, and fare are columns in the titanic dataset.

# Answer:->

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data
```

```python
# Create box plots using Seaborn
plt.figure(figsize=(12, 6))  # Adjust figure size if needed

# Box plot for 'age' across different 'pclass'
plt.subplot(1, 2, 1)  # Subplot 1 for age
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age Distribution across Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for 'fare' across different 'pclass'
plt.subplot(1, 2, 2)  # Subplot 2 for fare
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare Distribution across Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

# answer:->

```python
# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")
```

```python
diamonds_data
```

```python
# Create a histogram using Seaborn with 'cut' as hue
plt.figure(figsize=(8, 6))  # Set figure size

# Plot histogram with 'price' and differentiate by 'cut'
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30)

# Set plot title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

# Answer :->

```python
# Load the flights dataset
flights_data = sns.load_dataset("flights")
```

```python
flights_data
```

```python
# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")
```

```python
flights_pivot
```

```python
# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Set plot title and labels
plt.title('Passenger Numbers in Flights')
plt.xlabel('Year')
plt.ylabel('Month')
```
