```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title="Example of the Bokeh Plot", x_axis_label='x_axis', y_axis_label='y_axis')

# Add a circle glyph
p.circle(x, y, size=20, color="red", alpha=0.9)

# Display in Jupyter Notebook (if using)
output_notebook()

# Show the plot
show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
# Answer :->
```

Glyphs in Bokeh are visual shapes or markers used to represent data points on a plot. They can be circles, squares, lines, patches, or other geometric shapes that visually encode the data.

Here's an example showing how to add glyphs to a Bokeh plot:

```python
x = [1, 2, 3, 5, 4]
y = [4, 3, 5, 9, 6]

# Create a figure object
p = figure(title="Glyphs Example", x_axis_label='x-axis', y_axis_label='y-axis')

# Add different glyphs to the plot
p.circle(x, y, size=10, color="red", alpha=0.5)  # Circle glyph
p.square(x, y, size=12, color="green", alpha=0.7)  # Square glyph
p.line(x, y, line_width=2, color="navy", alpha=0.8)  # Line glyph

# Display in Jupyter Notebook (if using)
output_notebook()

# Show the plot
show(p)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
# Answer :->
```


Customizing the appearance of a Bokeh plot involves modifying various components such as axes, titles, legends, and more. Here's an example that demonstrates how you can customize these aspects:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend
```

```python
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [5, 4, 3, 2, 1]
```

```python
# Create a figure object
p = figure(title="Customized Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add glyphs with labels for the legend
circle = p.circle(x, y1, size=10, color="navy", alpha=0.5, legend_label="Circle")
line = p.line(x, y2, line_width=2, color="green", alpha=0.8, legend_label="Line")

# Customize axes
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_font_style = "italic"

# Customize title
p.title.text_color = "orange"
p.title.text_font_size = "18pt"
p.title.align = "center"

# Customize legend
p.legend.title = "Legend Title"
p.legend.title_text_font_style = "bold"
p.legend.label_text_font = "times"

# Display in Jupyter Notebook (if using)
output_notebook()

# Show the plot with the legend positioned at the top-left corner
p.add_layout(Legend(items=[("Circle", [circle]), ("Line", [line])]), 'above')
show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
```

```python
# Answer :->
```

Bokeh server is a component of the Bokeh library that enables the creation of interactive web applications and dashboards with live, reactive plots. It allows you to build complex, interactive visualizations that can update in real time based on user interactions, streaming data, or periodic updates.

To create an interactive plot using Bokeh server, you can follow these steps:

Define the Plot: Start by defining your plot, including the glyphs, data sources, and interactions you want to incorporate.

Create a Bokeh Application: Use the bokeh.application module to create a Python script that defines your application. This script should include a modify_doc() function that modifies the document (plot, layout, etc.) whenever changes occur.

Run the Bokeh Server: Run the Bokeh server using the bokeh serve command and provide the script file or directory containing the Python script as an argument.

Example: Let's create a simple Bokeh server application that updates a plot in real time:

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
import numpy as np
from bokeh.plotting import figure
```

```python
# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=[], y=[]))


# Create a figure
plot = figure(height=300, width=600, title="Real-time Data Update")
plot.line('x', 'y', source=source)

# Function to update the plot data
def update():
    new_data = dict(x=np.random.rand(100), y=np.random.rand(100))
    source.data = new_data

# Add periodic callback to update the plot every 200 milliseconds
curdoc().add_periodic_callback(update, 200)

# Add the plot to the current document
curdoc().add_root(plot)
```
