# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

##  Answer :->
### Matplotlib is a popular Python library used for creating static, interactive, and publication-quality visualizations. It provides a wide range of plotting capabilities and is extensively used in data analysis, scientific research, machine learning, and more.

### Matplotlib is used for several reasons, making it an essential tool in the Python ecosystem:
1. Data Visualization
2. Flexibility and Customization
3. Support for Various Plot Types 
4. Integration with Python Ecosystem
5. Cross-Platform Compatibility
##### Overall, Matplotlib's capabilities in creating customizable, high-quality plots, along with its versatility and integration within the Python ecosystem, make it a go-to tool for visualizing data in scientific research, data analysis, academic settings, and industry applications.


### The Pyplot module in Matplotlib offers a straightforward interface for creating various types of plots. Here are five commonly used plots that you can generate using Matplotlib's Pyplot:

1. Line Plot: Displays data points connected by straight line segments, commonly used to showcase trends over a continuous variable (e.g., time series data).

2. Scatter Plot: Represents individual data points as markers on a two-dimensional plane, useful for visualizing relationships between two numerical variables.

3. Histogram: Shows the distribution of a single numerical variable by dividing it into bins and displaying the frequency or count of data points within each bin.

4. Bar Plot: Displays categorical data with rectangular bars, often used to compare quantities or frequencies across different categories.

5. Pie Chart: Illustrates the proportion or percentage of each category in a dataset as a circular chart, useful for displaying relative sizes of categories.

#### These plots cover a range of visualization needs, allowing you to represent different types of data in an understandable and visually appealing manner using Matplotlib's Pyplot module.

### A scatter plot is a type of plot that displays values for two variables as points on a two-dimensional plane. Each point on the plot represents a single observation where the position of the point is determined by the values of the two variables - one plotted along the x-axis and the other plotted along the y-axis. Scatter plots are useful for visualizing relationships or correlations between two numerical variables.

### Here's an example code in Python to generate data for x and y variables and create a scatter plot using Matplotlib:



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y , c = '#1F2DEC',alpha=.9,marker='x')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("this is my x vs y")
plt.grid()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
### import numpy as np
### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# Answer :->
### The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to divide the figure into a grid and specify the position where each subplot will be placed.

### Here's an example drawing four line plots using the subplot() function with the provided data:



```python
import matplotlib.pyplot as plt
import numpy as np

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)  # (rows, columns, subplot number)
plt.plot(x, y1, 'r')  # Line 1 plot with red color
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')  # Line 2 plot with green color
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b')  # Line 3 plot with blue color
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'y')  # Line 4 plot with yellow color
plt.title('Line Plot 4')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

# Answer :->
### A bar plot is a type of chart that presents categorical data with rectangular bars. The length of each bar corresponds to the value it represents, making it easy to compare the magnitudes of different categories. Bar plots are commonly used to visualize and compare discrete categories or groups of data.

### Here's how you can create both a vertical bar plot and a horizontal bar plot using the provided data:

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')  # Creating a vertical bar plot
plt.title('Company Profits (Vertical Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.grid(axis='y')  # Show gridlines on the y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')  # Creating a horizontal bar plot
plt.title('company profits (Horizontal Bar Plot)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.grid(axis='x')  # Show gridlines on the x-axis
plt.tight_layout()
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
### box1 = np.random.normal(100, 10, 200)
### box2 = np.random.normal(90, 20, 200)

#Answer:-> 

### A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a graphical representation of the distribution of a dataset. It displays the five-number summary of a dataset: minimum, first quartile (Q1), median (second quartile, Q2), third quartile (Q3), and maximum. The box represents the interquartile range (IQR) between Q1 and Q3, while the whiskers extend to show the range of the data, excluding outliers.

### Box plots are used for several purposes:
1. Visualizing Spread and Skewness: They show the spread of the data and identify skewness or asymmetry in the distribution.
2. Identifying Outliers: They help in detecting outliers or extreme values in the dataset.
3. Comparing Distributions: Box plots allow for easy comparison of multiple datasets or groups.
#### Here's how you can create a box plot using the provided data:

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True, showmeans=True)
plt.title('Box Plot Comparison')
plt.xlabel('Box')
plt.ylabel('Values')
plt.xticks([1, 2], ['Box 1', 'Box 2'])  # Labeling x-axis ticks
plt.grid(axis='y')  # Show gridlines on the y-axis
plt.tight_layout()
plt.show()
```
